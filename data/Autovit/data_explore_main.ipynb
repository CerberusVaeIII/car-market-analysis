```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("car_data_autovit.csv")
```

```python
plt.style.use('ggplot')

x = df['mileage_km']
y = df['price_eur']

fig, ax = plt.subplots()

ax.scatter(x, y)
ax.set(xlim=(0, 400000), ylim=(0, 200000))
plt.xlabel('Mileage (km)')
plt.ylabel('Price (EUR)')
plt.title('Broad mileage vs price for Autovit.ro listings')
plt.show()
```

```python
avg_price_per_year = df.groupby('production_year')['price_eur'].mean()
avg_price_per_year.plot()
plt.xlabel("Production Year")
plt.ylabel("Average Price")
plt.title("Average Car Price Over Production Years")
plt.show()
```

```python
plt.style.use('ggplot')

x = df['engine_displacement_cm3']
y = df['price_eur']

fig, ax = plt.subplots()

ax.scatter(x, y)
ax.set(xlim=(0, 8000), ylim=(0, 200000))
plt.xlabel('Engine displacement (cm)')
plt.ylabel('Price (EUR)')
plt.title('Broad engine displacement vs price for Autovit.ro listings')
plt.show()
```

```python
avg_price_per_brand = df.groupby('brand')['price_eur'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['price_eur','power_hp','engine_displacement_cm3','mileage_km','production_year']].corr(), annot=True)
plt.show()
```

```python
top_brands = df['brand'].value_counts().head(5).index
sns.violinplot(x='brand', y='price_eur', data=df[df['brand'].isin(top_brands)])
```
