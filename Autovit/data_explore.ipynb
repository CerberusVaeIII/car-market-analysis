```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("car_data_autovit.csv")
```

```python
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
df['production_year'].plot(kind = "hist", bins = 20, edgecolor = "white")
plt.title("Distribution of car production years")
plt.xlabel("Production year")
plt.ylabel("Number of Cars")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['production_year'].plot(kind = "box", vert = False)
plt.title("Distribution of car production years")
plt.xlabel("Production year")
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
plot_df = df.dropna(subset=['engine_displacement_cm3', 'power_hp', 'price_eur'])

x = plot_df['engine_displacement_cm3']
y = plot_df['power_hp']
z = plot_df['price_eur']

ax.scatter3D(x, y, z, c='blue', marker='o')

ax.set_xlabel('Engine (cm³)')
ax.set_ylabel('Power (HP)')
ax.set_zlabel('Price (€)')
ax.set_title('3D Scatter Plot: Engine vs Power vs Price')

plt.show()
```
